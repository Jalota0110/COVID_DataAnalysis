```python
from geopy.geocoders import Nominatim
import time

geolocator = Nominatim(user_agent="geoapiExercises")

def get_continent(location):
    try:
        geo_info = geolocator.geocode(location)
        if geo_info:
            return geo_info.raw['address']['continent']
    except:
        return None
    return None

# Fill missing continent values
df['continent'] = df['continent'].fillna(df['location'].apply(lambda x: get_continent(x) if pd.notnull(x) else None))

print(df)
```
